```python
import numpy as np

n_x = 2                                 # 输入变量的维度。
m   = len(x_data)                       # 样本数。这里，x 和 y 的样本数量应相等。

x = np.array(x_data).reshape((m, n_x))  # 转为 numpy 类型的数组。
y = np.array(y_data).reshape((m, 1))    # 转为 numpy 类型的数组。
w = np.random.randn(n_x, 1)             # 初始化参数向量 w，且标准正态分布的随机函数，取值在 -1 到 1 之间。
b = np.random.randn(1, 1)               # 初始化参数 b，且标准正态分布的随机函数，取值在 -1 到 1 之间。

n_epoch = 100                            # 反向传播的迭代次数，也就是参数被更新的次数。
alpha   = 0.001                          # 学习率
```

```python
epsilon = 1e-9 # 代表一个无穷小的正数，对 log 函数进行修正。

def train_once(m, x, y, w, b, alpha):

    # 前向传播以计算成本函数值
    z = np.dot(x, w) + b
    a = 1 / (1 + np.exp(-z))
    J = - np.dot(y.T, np.log(a + epsilon)) - np.dot((1 - y).T, np.log(1 - a + epsilon)) # 由于 a 和 1 - a 有可能为 0，因此使用 epsilon 对 log(x) 函数进行修正，使 x 不为 0。

    # 使用链式求导法则求 J 对各个参数的导数。
    dz = a - y
    dw = np.dot(x.T, dz)
    db = dz.sum().reshape((1, 1))

    J /= m
    dw /= m
    db /= m

    print("J", J.flatten()) # flatten() 将多维向量展平为 1 维向量。
    print("w", w.flatten())
    print("b", b.flatten())
    print()

    w = w - alpha * dw
    b = b - alpha * db

    return w, b
```

```python
# Loop 以上 n_epoch 次，或直至成本函数 J 不再变小。
for i in range(n_epoch):
    w,b = train_once(m, x, y, w, b, alpha)
```
